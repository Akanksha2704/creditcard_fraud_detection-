# **data set information**
Out of 284,807 there are 492 frauds .The dataset is highly unbalanced and the positive class(frauds) account for 0.172% of all transcations.The input values are PCA transformations.the only features which have not been transformed with PCA are time and amount.feature "CLASS" is the variable and it takes 1 in case of fraud or 0 otherwise.
time: the time of transactions.
V1 - V28: various attributes of credit card transactions.
amount: amount involved in credit card transactions.
class: whether the given credit card transaction is fraudulent (1) or not (0).
  

# **Import modules**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

# **LOADING DATASET**

```python
df=pd.read_csv('creditcard.csv')
df.head()
```

```python
mean_value=df['V22'].mean()
mean_value1=df['V23'].mean()
mean_value2=df['V24'].mean()
mean_value3=df['V25'].mean()
mean_value4=df['V26'].mean()
mean_value5=df['V27'].mean()
mean_value6=df['V28'].mean()
mean_value7=df['Amount'].mean()
mean_value8=df['Class'].mean()
```

```python
df.isnull().sum()
```

```python
df['V22'].fillna(value=mean_value,inplace=True)
df['V23'].fillna(value=mean_value1,inplace=True)
df['V24'].fillna(value=mean_value2,inplace=True)
df['V25'].fillna(value=mean_value3,inplace=True)
df['V26'].fillna(value=mean_value4,inplace=True)
df['V27'].fillna(value=mean_value5,inplace=True)
df['V28'].fillna(value=mean_value6,inplace=True)
df['Amount'].fillna(value=mean_value7,inplace=True)
df['Class'].fillna(value=mean_value8,inplace=True)
```
